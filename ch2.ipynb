```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#(8)
#(a) College.csvをcollegeという名前で読み込む
college = pd.read_csv('./ISLR/College.csv')
college
```

```python
#(b) collegeの第1列にcollege_nameというカラム名をつける
college = college.rename(columns={college.columns[0] : 'college_name'})
# 第1列は別でもっておき、元のcollegeからは削除する
college_name = college.iloc[:,0]
college = college.iloc[:, 1:-1]
```

```python
#(c)(i)平均、最大、最小、分散、四分位、行数を出す
college.describe()
```

```python
#(c)(ii)最初の10列により、散布図行列を出す。
pd.plotting.scatter_matrix(college.iloc[:,1:10], figsize=(15, 15))
plt.show()
```

```python
#(c)(iii)OutstateとPrivateの箱ひげ図
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
college.boxplot(column="Outstate",by="Private",ax = ax)
plt.show()
```

```python
#(c)(iv)Top10perceから新たな変数Eliteを作る。Top10percがEnrollの50%を占める場合はEliteとみなす。Eliteの大学数と、OutstateとEliteの箱ひげ図を表示する
college['Elite'] = college[['Top10perc', 'Enroll']].apply(lambda x:  1 if (x[0] / x[1]) > 0.5 else 0, axis=1)
college.query('Elite == 1') # Eliteは二つある
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
college.boxplot(column="Outstate",by="Elite",ax = ax)
plt.show()
```

```python
#(c)(v)なんでもいいので量的変数でヒストグラムを作る。
fig, ax = plt.subplots(1, 5, figsize=(20, 5))
ax[0].hist(college["Outstate"])
ax[1].hist(college["Enroll"])
ax[2].hist(college["Top10perc"])
ax[3].hist(college["Top25perc"])
ax[4].hist(college["Accept"])
```

```python
#(c)(vi)データを詳しく調べ、どのような知見を得たか説明せよ
#(c)(ii)の散布図行列から得られた知見
#（見なくてもわかると思うが）入学願書の数、入学許可数、入学者数は各々正の相関がある
# 住居費食費と州外出身の学生の授業料には弱い正の相関がみられる
```

```python
#(10)
#(a)
# load_bostonはdeplicatedされる予定なので、警告が出るが無視でOK
from sklearn.datasets import load_boston
boston = load_boston()
#dataframeにする
X = boston["data"]
y = boston["target"]
feature_names = boston["feature_names"]
boston_df = pd.DataFrame(data=X, columns=feature_names)
boston_df['MEDV'] = pd.DataFrame(y)
boston_df.shape # 506行、13列ある
# カラムの意味などについては以下を参照のこと
# https://qiita.com/yut-nagase/items/6c2bc025e7eaa7493f89
```

```python
#(b)
pd.plotting.scatter_matrix(boston_df, figsize=(15, 15))
plt.show()
#相関がみられそうなものは以下
#NOXとAGE,NOXとDIS,DISとAGE,AGEとLSTAT
# (c)
#AGE、DIS、LSTAT、INDUSが関係していそう
```

```python
#(d)
# min,maxとmean、その他から考察する
# CRIMについては、maxが88という地域があり、平均から大きく離れている。ただ、ヒストグラムや箱ひげ図から、いくつかの値は外れ値と思われるため、これだけでは判断がつかない（平均が外れ値に引っ張られている）
# TAXは平均に対して、75パーセンタイルが666で、箱ひげ図から、外れ値がないと想定できるため、大きい地域があるといえる
# PTRATIOは75パーセンタイルが平均と近いため、高い地域はない（むしろ低い地域のほうが多い）
boston_df.boxplot(column=['CRIM', 'TAX', 'PTRATIO'])
# boston_df['CRIM'].plot.box()
# boston_df['TAX'].plot.box()
# boston_df['PTRATIO'].plot.box()
```
